```python
def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_type = df.dtypes
    mis_val_table = pd.concat([mis_val, mis_val_percent, mis_val_type], axis=1)
        
    mis_val_table_ren_columns = mis_val_table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values', 2: 'type'})
        
    mis_val_table_ren_columns = mis_val_table_ren_columns[ mis_val_table_ren_columns.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
        
    return mis_val_table_ren_columns
```

```python
missing_values = pd.DataFrame(missing_values_table(application))
missing_values
```

## Preprocessing

```python
application['OWN_CAR_AGE'].fillna(-1, inplace=True)
application['EXT_SOURCE_1'].fillna(0, inplace=True)
application['EXT_SOURCE_2'].fillna(0, inplace=True)
application['EXT_SOURCE_3'].fillna(0, inplace=True)
application['OCCUPATION_TYPE'].fillna("Unemployed", inplace=True)
```

```python
application['IS_EXT_SOURCE_1'] = application.EXT_SOURCE_1.isnull()
application['IS_EXT_SOURCE_2'] = application.EXT_SOURCE_2.isnull()
application['IS_EXT_SOURCE_3'] = application.EXT_SOURCE_3.isnull()
```

```python
application.drop(application.columns[100 * application.isnull().sum() / len(application) >= 40], axis=1, inplace=True)
```

```python
application.drop(columns=["AMT_REQ_CREDIT_BUREAU_HOUR", "AMT_REQ_CREDIT_BUREAU_DAY", "AMT_REQ_CREDIT_BUREAU_WEEK", "AMT_REQ_CREDIT_BUREAU_MON", "AMT_REQ_CREDIT_BUREAU_QRT", "AMT_REQ_CREDIT_BUREAU_YEAR", "NAME_CONTRACT_TYPE"], inplace=True)
```

```python
application.drop(columns=["DEF_30_CNT_SOCIAL_CIRCLE", "OBS_60_CNT_SOCIAL_CIRCLE", "DEF_60_CNT_SOCIAL_CIRCLE", "OBS_30_CNT_SOCIAL_CIRCLE"], inplace=True)
```

```python
application.dropna(axis=0, inplace=True)
```

```python
application = application[application["DAYS_BIRTH"] < 0]
application = application[application["DAYS_EMPLOYED"] < 0]
application = application[application["DAYS_REGISTRATION"] < 0]
application = application[application["DAYS_ID_PUBLISH"] < 0]
application = application[application["DAYS_LAST_PHONE_CHANGE"] < 0]
```

```python
reference_date = datetime(2019, 12, 31)
application['YEAR_OF_BIRTH'] = application['DAYS_BIRTH'].apply(lambda x: (reference_date + timedelta(days=x)).year)
application['YEAR_OF_EMPLOYMENT'] = application['DAYS_EMPLOYED'].apply(lambda x: (reference_date + timedelta(days=x)).year)
application['YEAR_OF_REGISTRATION'] = application['DAYS_REGISTRATION'].apply(lambda x: (reference_date + timedelta(days=x)).year)
application['YEAR_OF_ID_PUBLISH'] = application['DAYS_ID_PUBLISH'].apply(lambda x: (reference_date + timedelta(days=x)).year)
application['YEAR_OF_PHONE_CHANGE'] = application['DAYS_LAST_PHONE_CHANGE'].apply(lambda x: (reference_date + timedelta(days=x)).year)
```

```python
str_columns_application = application.select_dtypes(include=['object']).columns
application = pd.get_dummies(application, columns=str_columns_application)
```

```python
str_columns_previous = new_previous.select_dtypes(include=['object']).columns
previous = pd.get_dummies(new_previous, columns=str_columns_previous)
```

```python
l2 = previous.SK_ID_CURR.unique()
application = application[application['SK_ID_CURR'].isin(l2)]
```

### Train-test split

```python
ID_with_interest_rates = []
for i in range(previous.shape[0]):
    if not pd.isna(previous.iloc[i]["RATE_INTEREST_PRIMARY"]):
        ID_with_interest_rates.append(previous.iloc[i]["SK_ID_CURR"])
```

```python
final_test_df = previous[previous["SK_ID_CURR"].isin(ID_with_interest_rates)]
```

```python
new_indices = [id for id in l2 if id not in ID_with_interest_rates]
new_previous = previous[previous['SK_ID_CURR'].isin(new_indices)]
new_application = application[application['SK_ID_CURR'].isin(new_indices)]
```

```python
from sklearn.model_selection import train_test_split
half_1, half_2 = train_test_split(new_application, test_size=0.5, random_state=42)
train_default_application, test_default_application = train_test_split(half_1, test_size=0.2, random_state=42)
train_opti_application, test_opti_application = train_test_split(half_2, test_size=0.2, random_state=42)
test_opti_application = pd.concat([test_opti_application, final_test_df], axis=0)
```

```python
train_opti_previous = previous[previous['SK_ID_CURR'].isin(train_opti_application['SK_ID_CURR'])]
test_opti_previous = previous[previous['SK_ID_CURR'].isin(test_opti_application['SK_ID_CURR'])]
```

```python
X_train_default = train_default_application.drop(['TARGET'], axis=1)
y_train_default = train_default_application['TARGET']

X_test_default = test_default_application.drop(['TARGET'], axis=1)
y_test_default = test_default_application['TARGET']

columns = set(previous.columns) & set(application.columns)

Train_opti = train_opti_previous[list(columns)]

Test_opti = test_opti_previous[list(columns)]
```

```python
X_train_default.to_csv("X_train_default.csv")
y_train_default.to_csv("y_train_default.csv")

X_test_default.to_csv("X_test_default.csv")
y_test_default.to_csv("y_test_default.csv")

Train_opti.to_csv("Train_opti.csv")
Test_opti.to_csv("Test_opti.csv")
```
